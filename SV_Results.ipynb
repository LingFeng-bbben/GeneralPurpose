# Percent of no-interaction travel

Let's try to work out how often the robots were not being interacted with, as a percentage of their total (total distances that were recorded).

```python

# create a new data frame to avoid changing the original data
df2 = sim_data.copy()

df2["percent no interaction"] = ( df2["no interaction distance"] / (df2["dist followed"] + df2["no interaction distance"]) ) * 100

# Fix colours
colours = dict(zip(df2['Attempt'].unique(), sns.color_palette(n_colors=len(df2['Attempt'].unique()))))

bplot = sns.boxplot(hue='Attempt', y='percent no interaction', x='Condition', data=df2, width=0.4, palette=colours)

bplot.set_xticklabels(bplot.get_xticklabels(),rotation=30)
```
